### Plotting the .npy files sample code:

```python
import os
import numpy as np
import matplotlib.pyplot as plt

# Find all .npy files in the current directory
folder_path = '../data/npy_images'  # e.g., 'data/images'
npy_files = [f for f in os.listdir(folder_path) if f.endswith('.npy')]

# Optional: sort files if needed
npy_files.sort()

# Load only the first 50
npy_files = npy_files[:50]

# Visualize the first 5 .npy images
for i, file in enumerate(npy_files[:5]):
    file_path = os.path.join(folder_path, file)
    img = np.load(file_path)
    plt.figure()
    plt.imshow(img, cmap='gray')
    plt.title(f'Image: {file}')
    plt.axis('off')
plt.show()
```
